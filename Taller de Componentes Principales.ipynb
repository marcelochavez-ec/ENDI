## **Apuntes de Análisis Factorial y Componentes Principales**

### Sistema de Ecuaciones de Combinaciones Lineales (Componentes Principales)

El análisis de componentes principales (PCA) transforma las variables originales \( X_1, X_2, \dots, X_p \) en nuevas variables llamadas **componentes principales** \( PC_1, PC_2, \dots, PC_k \). Estas son combinaciones lineales de las variables originales:

$$
\begin{aligned}
    PC_1 &= w_{11}X_1 + w_{12}X_2 + \dots + w_{1p}X_p \\
    PC_2 &= w_{21}X_1 + w_{22}X_2 + \dots + w_{2p}X_p \\
    &\vdots \\
    PC_k &= w_{k1}X_1 + w_{k2}X_2 + \dots + w_{kp}X_p
\end{aligned}
$$

### Descripción de los términos:
- **\( PC_k \):** La \( k \)-ésima componente principal, una nueva variable que captura la mayor varianza posible de las variables originales en cada paso.
- **\( w_{ij} \):** El coeficiente o peso (**loading**) asociado a la \( j \)-ésima variable \( X_j \) en la \( i \)-ésima componente principal.
- **\( X_j \):** Las variables originales del conjunto de datos (atributos).

### Propiedades Clave:
1. Las componentes principales son ortogonales entre sí.
2. Cada componente explica una porción decreciente de la varianza total.
3. El objetivo del PCA es reducir la dimensionalidad manteniendo la mayor parte de la varianza.

Este sistema de ecuaciones describe cómo se calculan las componentes principales a partir de las variables originales, combinándolas con los pesos obtenidos de la descomposición de la matriz de covarianza o correlación.

### Varianza y Covarianza en Componentes Principales

La varianza explicada por cada componente principal se calcula como:

$$
\text{Var}(PC_k) = \sum_{j=1}^p w_{kj}^2 \cdot \text{Var}(X_j) + \sum_{j \neq j'} w_{kj} w_{kj'} \cdot \text{Cov}(X_j, X_{j'})
$$

La covarianza entre dos componentes principales \( PC_k \) y \( PC_l \) es:

$$
\text{Cov}(PC_k, PC_l) = \sum_{j=1}^p \sum_{j'=1}^p w_{kj} w_{lj'} \cdot \text{Cov}(X_j, X_{j'})
$$

### Descripción:
\( \text{Var}(PC_k) \): Varianza del componente principal \( k \), que mide cuánta varianza del conjunto original captura \( PC_k \).
\( \text{Cov}(PC_k, PC_l) \): Covarianza entre los componentes \( k \) y \( l \).
\( w_{kj} \): Coeficiente del \( j \)-ésimo atributo en el componente \( PC_k \).
\( \text{Var}(X_j) \): Varianza de la \( j \)-ésima variable original.
\( \text{Cov}(X_j, X_{j'}) \): Covarianza entre las variables originales \( X_j \) y \( X_{j'} \).

### Propiedades Clave:
1. La covarianza entre componentes principales es cero (\( \text{Cov}(PC_k, PC_l) = 0 \)) debido a la ortogonalidad.
2. Las componentes principales son combinaciones lineales que maximizan la varianza.
